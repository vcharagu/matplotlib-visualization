```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("marketing raw data.csv")
df
```

```python
plt.plot(df["Visitors"])
plt.show()
```

```python
plt.plot(df["Visitors"])
plt.xticks(range(0,len(df["Visitors"])),df['Date'])
plt.show()
```

```python
a=df.tail(10).reset_index()
plt.plot(a["Visitors"])
plt.xticks(range(0,len(a["Visitors"])),a['Date'],rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.bar(range(0,len(a["Visitors"])),a['Visitors'],label="No of Vizs",linewidth=10,color="red")
plt.plot(a["Revenue"])
plt.xticks(range(0,len(a["Visitors"])),a['Date'],rotation=90)
plt.title("Revenue Per Day")
plt.legend(loc="upper center")
plt.xlabel("Dates")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.bar(range(0,len(a["Revenue"])),a['Revenue'],label="No of Vizs",linewidth=10,color="green")
plt.bar(range(0,len(a["Visitors"])),a['Marketing Spend'],label="Marketing Spend	",linewidth=10,color="purple")

plt.plot(a["Visitors"])
plt.xticks(range(0,len(a["Visitors"])),a['Date'],rotation=90)
plt.title("Revenue Per Day")
plt.legend(loc="upper center")
plt.xlabel("Dates")
plt.show()
```

```python
a[["Revenue","Marketing Spend"]].plot(kind="bar",color=["green","red"],figsize=(12,8))
a['Visitors'].plot(c="blue",linestyle="-.",secondary_y=True)
plt.xticks(range(0,len(a["Visitors"])),a['Date'],rotation=90)
plt.title("Revenue Per Day")
plt.legend(loc="upper center")
plt.xlabel("Dates")
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['Visitors'],df['Marketing Spend'],c="red",marker="+",label="marketing")
plt.title("Visitors and Revenue Per Day")
plt.legend(loc="upper left")
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```

```python
groups=df[["Promo","Revenue","Marketing Spend"]].groupby("Promo")
groups
fig, ax=plt.subplots(figsize=(15,10))

for promo, group in groups:
  ax.scatter(group["Revenue"],group['Marketing Spend'],marker="x",label=promo)
ax.legend()
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")
plt.title("Relationship b2in marketing spend and Revenue by promo",fontsize=20)
```
